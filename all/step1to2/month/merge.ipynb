```python
import pandas as pd
import glob
import os

MONTH_DIR  = './data/'
OUTPUT_DIR = './data/'

# products 합치기
product_files = glob.glob(MONTH_DIR + '*_products_*_2026-04.csv')
df_products = pd.concat([pd.read_csv(f) for f in sorted(product_files)], ignore_index=True)
print(f'products 합계: {len(df_products):,}건')
print(f'파일 목록: {[os.path.basename(f) for f in sorted(product_files)]}')

# reviews 합치기
review_files = glob.glob(MONTH_DIR + '*_reviews_*_2026-04.csv')
df_reviews = pd.concat([pd.read_csv(f) for f in sorted(review_files)], ignore_index=True)
print(f'\nreviews 합계: {len(df_reviews):,}건')
print(f'파일 목록: {[os.path.basename(f) for f in sorted(review_files)]}')

# 저장
df_products.to_csv(OUTPUT_DIR + '4m_products.csv', index=False, encoding='utf-8-sig')
df_reviews.to_csv(OUTPUT_DIR + '4m_reviews.csv', index=False, encoding='utf-8-sig')

print(f'\n저장 완료')
print(f'  4m_products.csv: {len(df_products):,}건, 컬럼: {df_products.columns.tolist()}')
print(f'  4m_reviews.csv:  {len(df_reviews):,}건, 컬럼: {df_reviews.columns.tolist()}')
```

```python
total = 0
for f in sorted(review_files):
    df_temp = pd.read_csv(f)
    print(f'  {os.path.basename(f)}: {len(df_temp):,}건')
    total += len(df_temp)
print(f'\n합계: {total:,}건')
```
